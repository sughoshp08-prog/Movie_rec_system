```python
import pandas as pd
import faiss
import requests
import subprocess
import numpy as np
```

```python
df = pd.read_csv(r"netflix_titles.csv")
```

```python
df.head()
```

```python
def txt_rep(row):
    text_rep = f"""Type: {row['type']},
Title: {row['title']},
Director: {row['director']},
Cast: {row['cast']},
Released: {row['release_year']},
Genres: {row['listed_in']},
Description: {row['description']}"""
    return text_rep
```

```python
df['text_rep'] = df.apply(txt_rep, axis = 1)
```

```python
df['text_rep'].values[0]
```

```python
dim = 768
index = faiss.IndexFlatL2(dim)
X = np.zeros((len(df['text_rep']), dim), dtype='float32')
```

```python

def start_ollama():
    try:
        requests.get("http://localhost:11434")
    except:
        subprocess.Popen(["ollama", "serve"])

start_ollama()





for i, rep in enumerate(df['text_rep']):
    if i%100==0:
        print(i)
    res = requests.post('http://localhost:11434/api/embeddings',
                       json = {
                           'model': 'nomic-embed-text',
                           'prompt': rep
                       }
                       )
    embeddings = res.json()['embedding']
    X[i] = np.array(embeddings)

index.add(X)
```
